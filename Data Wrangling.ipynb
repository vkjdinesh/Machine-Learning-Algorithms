```python
#1. Load the dataset
dataset = pd.read_csv("https://raw.githubusercontent.com/datasets/covid-19/main/data/time-series-19-covid-combined.csv")
dataset.head(5)
```

```python
#dimension : rows, columns
dataset.shape
```

```python
#describe the dataset
dataset.describe()
```

```python
#Check the variable(Numerical / categorical)
numerical_variable = dataset.columns[dataset.dtypes !=  'object']
dataset[numerical_variable].head(5)

categorical_variable = dataset.columns[dataset.dtypes ==  'object']
dataset[categorical_variable].head(5)
```

```python
dataset[numerical_variable].head(5)
```

```python
dataset[categorical_variable].head(5)
```

```python
#Missing values
dataset.isnull().head(5)
dataset[numerical_variable].isnull().head(5)
dataset[categorical_variable].isnull().head(5)

#False - No missing values
#true = we have the missing values
```

```python
#Find out the the percentage of missing values
len(dataset)
```

```python
dataset.isnull().sum()/len(dataset)
dataset[numerical_variable].isnull().sum()
dataset[numerical_variable].isnull().sum()/len(dataset)
```

```python
#Replace
dataset.fillna('KIT')
```

```python
dataset['Confirmed'].replace(0, 1)
dataset.replace(0, "KIT")
dataset['Confirmed'].replace([0, 1],['KIT',0]).head(15)
```

```python
#Create the rows and add it with the original dataset
#values of variable, instructions about the variable
row_insert = pd.Series(['2022-05-20','Cambodia','TTY', 2000, 500, 0 ], index = ['Date', 'Country/Region', 'Province/State', 'Confirmed', 'Recovered', 'Deaths'])
dataset.append(row_insert, ignore_index=True)
```

```python
#call the row
#datasetname.iloc[] - we can call the specific row based on the index
dataset.iloc[0]
dataset.iloc[1:5]
dataset.iloc[1:6]
dataset.iloc[ :7]
```

```python
#loc - to cal the specific values, 
#index
dataset = dataset.set_index(dataset['Country/Region'])
dataset.loc['India']
```

```python
#rename
dataset.rename(columns={'Country/Region' : 'Country'}).head(5)
```

```python
print('Sum:', dataset['Deaths'].max())
dataset
```

```python
#delete col
dataset.drop('Province/State', axis=1).head(5)
dataset.drop(dataset.columns[1], axis=1)
```
